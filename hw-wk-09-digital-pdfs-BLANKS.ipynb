```python
import pandas as pd
import tabula
import glob
```

```python
## we need to make the pdfs into a list we can iterate through
pdfs = glob.glob("downloads/*.pdf")
```

```python
pdfs
```

```python
## now we need to read all the pages in each file and put each
#individual table within a list
all_tables = []
for pdf in pdfs:
    tables = tabula.read_pdf(pdf, pages="all")
    for table in tables: 
        all_tables.append(table)
        
all_tables
    

```

```python
## copy the function to combine tables

def combine_tables(list_name, csv_name):
    '''
    takes multiple dataframes in a list, concats them
    exports to .csv
    parameter1: list name
    parameter2: csv file name including.csv
    '''
    df = pd.concat(list_name)
    df.reset_index(inplace = True, drop = True)
    df.to_csv(csv_name, encoding = "UTF-8", index = False)
    return df
```

```python
fuds_df = combine_tables(all_tables, "fuds.csv")
```
